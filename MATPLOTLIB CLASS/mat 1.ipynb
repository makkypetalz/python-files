```python
temp = np.arange(90,160)
print(temp)
```

```python
df = pd.DataFrame(temp, columns=['Fahrenheit'])
print(df)
```

```python
def bvfgff(fto):
    return ((fto-32)*5/9)
```

```python
df['celsius'] = df['Fahrenheit'].apply(bvfgff)
df
```

```python
plt.plot(df['celsius'], df['Fahrenheit'])
```
