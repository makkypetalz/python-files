# How to add a legend

```python
stock1 = [200,400,500]
stock2 = [300,500,600]
stock3 = [400,700,800]
```

```python
plt.plot(stock1)
plt.plot(stock2)
plt.plot(stock3)
```

```python
plt.plot(stock1, label = 'MTN')
plt.plot(stock2, label = 'GLO')
plt.plot(stock3, label = '9MOBILE')
plt.legend()
```

```python
plt.plot(stock1, label = 'MTN')
plt.plot(stock2, label = 'GLO')
plt.plot(stock3, label = '9MOBILE')
plt.legend(loc = 'lower left')
```

```python
y = np.arange(100)
fig, axs = plt.subplots(2,2)

axs[0,0].plot(y, np.sin(y))
axs[0,0].set_title('sine wive')

axs[0,1].plot(y, np.cos(y))
axs[0,1].set_title('cosine wive')

axs[1,0].plot(y, np.random.random(100))
axs[1,0].set_title('cosine wive')

axs[1,1].plot(y, np.log(y))
axs[1,1].set_title('log function')
```

# 3Dplot

```python
ax = plt.axes(projection = '3d')
a = np.random.random(100)
b = np.random.random(100)
c = np.random.random(100)
ax.scatter(a,b,c)
ax.set_title('SCORES')
```

```python
ax = plt.axes(projection = '3d')
a = np.arange(0,100,0.1)
b = np.sin(a)
c = np.cos(a)
ax.scatter(a,b,c)
```

```python
ax = plt.axes(projection = '3d')
a = np.arange(0,100,0.1)
b = np.arange(0,100,0.1)
c = np.cos(a+b)
ax.scatter(a,b,c)
```
